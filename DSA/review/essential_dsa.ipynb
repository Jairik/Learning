## 1) Array (High)

Use arrays/lists for contiguous storage and index-based access. In interviews, common moves are two-pointers, prefix/suffix precompute, and sliding window.

- Strength: O(1) random access
- Cost: inserts/deletes in middle are O(n)

```python
def two_sum(nums, target):
    seen = {}
    for i, x in enumerate(nums):
        if target - x in seen:
            return [seen[target - x], i]
        seen[x] = i
    return []


def max_subarray(nums):  # Kadane's algorithm
    best = cur = nums[0]
    for x in nums[1:]:
        cur = max(x, cur + x)
        best = max(best, cur)
    return best

print(two_sum([2, 7, 11, 15], 9))
print(max_subarray([-2, 1, -3, 4, -1, 2, 1, -5, 4]))
```

## 2) String (High)

Strings are array-like but immutable in Python. Frequent interview patterns: two-pointers, frequency counts, parsing/state machine.

```python
from collections import Counter


def is_palindrome(s):
    filtered = [c.lower() for c in s if c.isalnum()]
    l, r = 0, len(filtered) - 1
    while l < r:
        if filtered[l] != filtered[r]:
            return False
        l += 1
        r -= 1
    return True


def is_anagram(a, b):
    return Counter(a) == Counter(b)

print(is_palindrome("A man, a plan, a canal: Panama"))
print(is_anagram("listen", "silent"))
```

## 3) Hash Table (Mid)

Use hash maps/sets for expected O(1) lookup/update. Core for dedup, membership checks, counting, and index mapping.

```python
from collections import defaultdict


def group_anagrams(words):
    groups = defaultdict(list)
    for w in words:
        key = tuple(sorted(w))
        groups[key].append(w)
    return list(groups.values())


def first_unique_char(s):
    freq = defaultdict(int)
    for c in s:
        freq[c] += 1
    for i, c in enumerate(s):
        if freq[c] == 1:
            return i
    return -1

print(group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"]))
print(first_unique_char("leetcode"))
```

## 4) Recursion (Mid)

Recursion shines on trees/graphs/divide-and-conquer/backtracking. Focus on base case, transition, and stack depth.

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)


def subsets(nums):
    res = []

    def dfs(i, path):
        if i == len(nums):
            res.append(path[:])
            return
        dfs(i + 1, path)          # skip
        path.append(nums[i])      # take
        dfs(i + 1, path)
        path.pop()

    dfs(0, [])
    return res

print(factorial(5))
print(subsets([1, 2, 3]))
```

## 5) Sorting and Searching (High)

Sorting enables binary-search-style reasoning and order-based sweeps. Searching is often binary search on index or answer space.

```python
def binary_search(nums, target):
    l, r = 0, len(nums) - 1
    while l <= r:
        m = (l + r) // 2
        if nums[m] == target:
            return m
        if nums[m] < target:
            l = m + 1
        else:
            r = m - 1
    return -1


def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    merged = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i]); i += 1
        else:
            merged.append(right[j]); j += 1
    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged

print(binary_search([1, 3, 4, 7, 10], 7))
print(merge_sort([5, 2, 4, 1, 3]))
```

## 6) Matrix (High)

Matrix problems usually map to index transforms, directional traversal, or grid graph search (BFS/DFS).

```python
def rotate_90_clockwise(matrix):
    n = len(matrix)
    for i in range(n):
        for j in range(i + 1, n):
            matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
    for row in matrix:
        row.reverse()
    return matrix


def num_islands(grid):
    if not grid:
        return 0
    rows, cols = len(grid), len(grid[0])

    def dfs(r, c):
        if r < 0 or c < 0 or r >= rows or c >= cols or grid[r][c] != "1":
            return
        grid[r][c] = "0"
        dfs(r + 1, c); dfs(r - 1, c); dfs(r, c + 1); dfs(r, c - 1)

    count = 0
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == "1":
                count += 1
                dfs(r, c)
    return count

print(rotate_90_clockwise([[1,2,3],[4,5,6],[7,8,9]]))
print(num_islands([list("110"), list("010"), list("011")]))
```

## 7) Linked List (Mid)

Linked lists trade random access for cheap pointer rewiring. Interview staples: reversal, cycle detection, merge.

```python
class ListNode:
    def __init__(self, val=0, nxt=None):
        self.val = val
        self.next = nxt


def reverse_list(head):
    prev, cur = None, head
    while cur:
        nxt = cur.next
        cur.next = prev
        prev = cur
        cur = nxt
    return prev


def to_list(head):
    out = []
    while head:
        out.append(head.val)
        head = head.next
    return out

h = ListNode(1, ListNode(2, ListNode(3)))
print(to_list(reverse_list(h)))
```

## 8) Queue (Mid)

Queue = FIFO; useful for BFS, scheduling, buffering. In Python, use `collections.deque` for O(1) append/pop from ends.

```python
from collections import deque


def bfs_levels(tree):
    if not tree:
        return []
    q = deque([tree])
    levels = []
    while q:
        size = len(q)
        level = []
        for _ in range(size):
            node = q.popleft()
            level.append(node.val)
            if node.left: q.append(node.left)
            if node.right: q.append(node.right)
        levels.append(level)
    return levels
```

## 9) Stack (Mid)

Stack = LIFO; ideal for nested structure parsing, backtracking state, monotonic stack problems.

```python
def valid_parentheses(s):
    pairs = {')': '(', ']': '[', '}': '{'}
    st = []
    for ch in s:
        if ch in '([{':
            st.append(ch)
        elif ch in pairs:
            if not st or st.pop() != pairs[ch]:
                return False
    return not st


class MinStack:
    def __init__(self):
        self.st = []
        self.min_st = []

    def push(self, x):
        self.st.append(x)
        if not self.min_st or x <= self.min_st[-1]:
            self.min_st.append(x)

    def pop(self):
        x = self.st.pop()
        if x == self.min_st[-1]:
            self.min_st.pop()
        return x

    def get_min(self):
        return self.min_st[-1]

print(valid_parentheses("{[()]}"))
ms = MinStack(); ms.push(3); ms.push(1); ms.push(2); ms.pop(); print(ms.get_min())
```

## 10) Tree (High)

Trees encode hierarchy. Most interview tasks reduce to traversal order (DFS/BFS), subtree aggregation, or BST invariants.

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def inorder(root):
    if not root:
        return []
    return inorder(root.left) + [root.val] + inorder(root.right)


def max_depth(root):
    if not root:
        return 0
    return 1 + max(max_depth(root.left), max_depth(root.right))


root = TreeNode(4, TreeNode(2, TreeNode(1), TreeNode(3)), TreeNode(6))
print(inorder(root))
print(max_depth(root))
```

## 11) Graph (High)

Graphs model arbitrary relationships. Core interview toolkit: adjacency list, BFS/DFS traversal, cycle detection, and shortest path.

```python
from collections import defaultdict, deque


def bfs_graph(adj, start):
    seen = {start}
    q = deque([start])
    order = []
    while q:
        u = q.popleft()
        order.append(u)
        for v in adj[u]:
            if v not in seen:
                seen.add(v)
                q.append(v)
    return order


def has_cycle_directed(adj):
    visiting, visited = set(), set()

    def dfs(u):
        if u in visiting:
            return True
        if u in visited:
            return False
        visiting.add(u)
        for v in adj[u]:
            if dfs(v):
                return True
        visiting.remove(u)
        visited.add(u)
        return False

    return any(dfs(node) for node in list(adj.keys()))


g = defaultdict(list, {0: [1, 2], 1: [2], 2: [3], 3: []})
print(bfs_graph(g, 0))
print(has_cycle_directed(g))
```

## 12) Heap (Mid)

Heap gives fast access to min/max element (Python `heapq` is min-heap). Useful for top-k, scheduling, stream problems.

```python
import heapq


def kth_largest(nums, k):
    heap = nums[:k]
    heapq.heapify(heap)
    for x in nums[k:]:
        if x > heap[0]:
            heapq.heapreplace(heap, x)
    return heap[0]


print(kth_largest([3, 2, 1, 5, 6, 4], 2))
```

## 13) Trie (Mid)

Trie is a prefix tree for string keys. Strong when prefix queries dominate and alphabet is bounded.

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False


class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for ch in word:
            node = node.children.setdefault(ch, TrieNode())
        node.is_end = True

    def search(self, word):
        node = self.root
        for ch in word:
            if ch not in node.children:
                return False
            node = node.children[ch]
        return node.is_end

    def starts_with(self, prefix):
        node = self.root
        for ch in prefix:
            if ch not in node.children:
                return False
            node = node.children[ch]
        return True


tr = Trie()
tr.insert("apple")
print(tr.search("apple"), tr.search("app"), tr.starts_with("app"))
```

## 14) Interval (Mid)

Interval problems are often sort-then-sweep. Key ideas: overlap conditions and maintaining current merged range.

```python
def merge_intervals(intervals):
    intervals.sort(key=lambda x: x[0])
    merged = []
    for s, e in intervals:
        if not merged or s > merged[-1][1]:
            merged.append([s, e])
        else:
            merged[-1][1] = max(merged[-1][1], e)
    return merged


print(merge_intervals([[1,3],[2,6],[8,10],[15,18]]))
```

## 15) Dynamic Programming (Low)

DP = optimize overlapping subproblems with memoization/tabulation. Interview framing: define state, recurrence, base cases.

```python
def coin_change(coins, amount):
    dp = [amount + 1] * (amount + 1)
    dp[0] = 0
    for x in range(1, amount + 1):
        for c in coins:
            if c <= x:
                dp[x] = min(dp[x], dp[x - c] + 1)
    return -1 if dp[amount] == amount + 1 else dp[amount]


print(coin_change([1, 2, 5], 11))
```

## 16) Binary / Bit Manipulation (Low)

Bit tricks give compact O(1) operations for flags/parity/counts; frequent in constraints-heavy tasks.

```python
def hamming_weight(n):
    count = 0
    while n:
        n &= n - 1
        count += 1
    return count


def single_number(nums):
    x = 0
    for n in nums:
        x ^= n
    return x

print(hamming_weight(0b101101))
print(single_number([4, 1, 2, 1, 2]))
```

## 17) Math (Low)

Common interview math: gcd/lcm, modular arithmetic, combinatorics, primes, and fast exponentiation.

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a


def fast_pow(x, n):
    res = 1
    while n > 0:
        if n & 1:
            res *= x
        x *= x
        n >>= 1
    return res

print(gcd(48, 18))
print(fast_pow(2, 10))
```

## 18) Geometry (Low)

Geometry tasks usually rely on vector/cross-product reasoning, orientation tests, and robust boundary handling.

```python
def orientation(a, b, c):
    # >0: counterclockwise, <0: clockwise, =0: collinear
    return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])


def rectangle_area(x1, y1, x2, y2):
    return abs(x2 - x1) * abs(y2 - y1)

print(orientation((0, 0), (2, 0), (1, 1)))
print(rectangle_area(0, 0, 3, 4))
```
