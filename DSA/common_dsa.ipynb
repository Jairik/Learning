## **Linked List**

A list of interconnected nodes, connected by pointers. Can be sorted or unsorted.

#### Time Complexities
- Search: O(n)
- Insert (Unsorted): O(1)
- Insert (Sorted): O(n)
- Delete: O(n)


#### Use Cases

- LRU (Least Recently Used) Caches
- Undo/Redo Stacks
- Memory Efficient with Unknown Size

#### Common Problems

- Reverse a Linked List
- Merge Two Sorted Lists
- Remove Duplicates
- Cycle Detection
- Swap Nodes in Pairs
- Intersection of Two Linked Lists
- Reverse Nodes

#### Key Techniques & Concepts

- Two-pointer approaches
- Using dummy nodes
- Recursion vs. iteration
- Hash Maps (for cycle detection/intersection/deduplication)

#### Arrays vs Linked Lists

- Arrays have better cache locality and faster access patterns, although could be worse for unknown sizes
- Arrays have simpler indexing (potentially 0(1))
- Arrays have less memory overhead

```python
''' Instantiate a linked list class '''
class Node:
    def __init__(self, data):
        self.data = data
        self.next_node = None

# Unsorted linked list implementation
class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0
    
    # O(1) time complexity
    def insert(self, data):
        new_node = Node(data)
        new_node.next_node = self.head
        self.head = new_node
        self.size += 1
    
    # O(n) time complexity
    def delete(self, data):
        current_node = self.head
        previous_node = None
        
        while current_node is not None:
            if current_node.data == data:
                if previous_node is None:
                    self.head = current_node.next_node
                else:
                    previous_node.next_node = current_node.next_node
                self.size -= 1
                return
            previous_node = current_node
            current_node = current_node.next_node
    
    # O(n) time complexity
    def search(self, data):
        current_node = self.head
        
        while current_node is not None:
            if current_node.data == data:
                return True
            current_node = current_node.next_node
            
        return False
    
    # O(n) time complexity
    def display(self):
        current_node = self.head
        elements = []
        
        while current_node is not None:
            elements.append(current_node.data)
            current_node = current_node.next_node
            
        print("Linked List:", " -> ".join(map(str, elements)))
```

## **Hashmaps**
An O(1) key value store. Uses a hash function to map values to indexes. Implemented as a python dict.

#### Time Complexities
- Insert: O(1), O(n) worst case with hash collisions
- Delete: O(1), O(n) worst case with hash collisions
- Search/Lookup: O(1), O(n) worst case with hash collisions
- Access by Key: O(1)
- Iteration over all entries: O(n)
- Clear/Empty: O(n)

#### Use Cases
- Caching/Memoization
- Counting frequencies
- Deduplication
- Lookup tables
- Two-sum problems
- Graph adjacency lists
- LRU/LFU Caches (Least Recently Used, Least Frequently Used)
- Grouping Data (by some property)

#### Common Problems
- Two Sum
- Valid Anagram
- Longest Substring Without Repeating Characters
- Majority Element
- First Unique Character in String
- Intersection of Two Arrays
- Cycle Detection
- Word Patterns

#### Key Techniques & Concepts
- Sliding Window + Hashmap
- Two-Pass Approach (build the hashmap in first pass, use lookups for second pass)
- Complement/Inverse Lookup
- Frequency Counting
- Index Mapping (solve distance/gap problems efficiently)
- Collision Handling (chaining vs open addressing, python uses chaining)
- Hashmap vs Hashset (Sets just track existence, maps track key-value pairs)

#### Optimizations When Using HashMaps
- Initialize with expected size to reduce rehashing
- Use hash sets for existence checks (faster than maps)
- Be aware of hash function quality for custom objects

```python
hashmap: dict = {}  # Instatiate the hashmap

def insert(key, value):
    hashmap[key] = value  # Inserting (O(1) time complexity)
    
def delete(key):
    # Deleting (O(1) time complexity)
    if key in hashmap:
        del hashmap[key]
        
def search(key):
    return hashmap.get(key, None)  # Searching (O(1) time complexity)  
```

## **Binary Search**

An **O(logn)** method of searching a sorted array.

#### Use Cases
- Fast lookup in sorted data
- Lower/upper bounds (finding insertion positions or ranges)
- Optimizations

#### Common Problems
- Search Insert Position
- Find First and Last Position of Elemenent
- Search in Rotated Sorted Array
- Find Minimum in Rotated Sorted Array
- Peak Element
- Find kth smallest/largest in sorted matrix
- Median of two sorted arrays (binary search on partitions)
- Capacity to ship packages in D Days (binary search on answer)
- Minimize max distance to gas station (binary search on answer)

#### Key Techniques & Concepts
- Invariant thinking (describe what low and high represent)
- Use mid = low + (high - low) // 2 to avoid overflow, while low <= high for exact match search and while low < high for boundary search

```python
# Binary search implementation (O(log n) time complexity), returns index of element if found (looking for exact match)
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    
    # Would modify some of this logic if looking for bounds
    while left <= right:
        mid = left + (right - left) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
            
    return -1  # Target not found
```

## **Trees**

A hierarchical data structure with a root node and child subtrees. Common variants include binary trees, BSTs, heaps, and tries.

#### Time Complexities
- Traversals (BFS/DFS): O(n)
- Search/Insert/Delete in balanced BST: O(log n)
- Search/Insert/Delete in unbalanced BST: O(n)

#### Use Cases
- Hierarchical data (file systems, org charts, DOM)
- Sorted data with fast range queries (BST)
- Priority queues and scheduling (heap)
- Autocomplete and prefix matching (trie)

#### Common Problems
- Preorder/Inorder/Postorder Traversal
- Level Order Traversal (BFS)
- Max/Min Depth, Diameter
- Validate BST / Kth Smallest in BST
- Lowest Common Ancestor
- Path Sum / Root-to-Leaf Paths
- Serialize/Deserialize Tree
- Invert/Mirror Tree

#### Key Techniques & Concepts
- Recursion with base cases (divide and conquer)
- Iterative DFS with stack, BFS with queue
- Subtree invariants (e.g., BST min/max bounds)
- Use of parent pointers or backtracking when needed

```python
# BST implementation
class TreeNode:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
        
class BinarySearchTree:
    def __init__(self):
        self.root = None
    
    def insert(self, key):
        if self.root is None:
            self.root = TreeNode(key)
        else:
            self._insert_recursive(self.root, key)
    
    def _insert_recursive(self, node, key):
        if key < node.val:
            if node.left is None:
                node.left = TreeNode(key)
            else:
                self._insert_recursive(node.left, key)
        else:
            if node.right is None:
                node.right = TreeNode(key)
            else:
                self._insert_recursive(node.right, key)
    
    def search(self, key):
        return self._search_recursive(self.root, key)
    
    def _search_recursive(self, node, key):
        if node is None:
            return False
        if node.val == key:
            return True
        elif key < node.val:
            return self._search_recursive(node.left, key)
        else:
            return self._search_recursive(node.right, key)
```

## **Stacks/Queues**

Linear data structures with restricted access patterns.

#### Time Complexities
- Stack push/pop: O(1)
- Queue enqueue/dequeue: O(1)
- Search: O(n)

#### Use Cases
- Parsing expressions / balancing parentheses (stack)
- Undo/redo and backtracking (stack)
- BFS traversal and scheduling (queue)
- Producer/consumer buffers (queue)

#### Common Problems
- Valid Parentheses
- Next Greater Element
- Min Stack / Max Stack
- Sliding Window Maximum (deque)
- Implement Queue using Stacks / Stack using Queues

#### Key Techniques & Concepts
- LIFO vs FIFO behavior
- Use `deque` for O(1) queue operations
- Monotonic stack/queue for range queries

```python
from collections import deque

# Stack (LIFO) using list
stack = []
stack.append(10)  # push
stack.append(20)
top = stack.pop()  # pop -> 20

# Queue (FIFO) using deque
queue = deque()
queue.append(10)  # enqueue
queue.append(20)
front = queue.popleft()  # dequeue -> 10
```

## **Heaps**

- Top K / Kth largest/smallest
- Merge K sorted lists
- Dijkstra and Prim's algorithms

#### Common Problems
- Kth Largest Element
- Find Median from Data Stream
- Task Scheduler
- Reorganize String
- Top K Frequent Elements

#### Key Techniques & Concepts
- Index math in array heap (parent/children)
- Use `heapq` in Python (min-heap)
- Simulate max-heap with negative values

```python
import heapq

# Min-heap
min_heap = []
heapq.heappush(min_heap, 5)
heapq.heappush(min_heap, 2)
heapq.heappush(min_heap, 8)
smallest = heapq.heappop(min_heap)  # -> 2

# Max-heap (store negatives)
max_heap = []
heapq.heappush(max_heap, -5)
heapq.heappush(max_heap, -2)
heapq.heappush(max_heap, -8)
largest = -heapq.heappop(max_heap)  # -> 8
```

## **Dynamic Programming**

An optimization technique that solves problems by combining solutions to overlapping subproblems.

#### Time Complexities
- Often reduces exponential recursion to polynomial time
- Space depends on memoization/table size (O(n) or O(n*m))

#### Use Cases
- Optimization problems (min/max, count ways)
- Sequence alignment and edit distance
- Knapsack and coin change
- Longest increasing subsequence / LCS

#### Common Problems
- Fibonacci / Climbing Stairs
- Coin Change / Coin Change II
- 0/1 Knapsack
- Longest Common Subsequence
- House Robber
- Partition Equal Subset Sum

#### Key Techniques & Concepts
- Top-down memoization vs bottom-up tabulation
- State definition and transitions
- Base cases and iteration order
- Space optimization (rolling arrays)

```python
# Bottom-up Fibonacci (tabulation)
def fib(n):
    if n <= 1:
        return n
    dp = [0] * (n + 1)
    dp[1] = 1
    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
    return dp[n]

# Example
fib_10 = fib(10)  # -> 55
```
